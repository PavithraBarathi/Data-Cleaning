```python
#Write python code for indexing rows and columns based on the following conditions:
#Assume we have the following dataframe:
#data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],
#'age': [25, 35, 40, 28],
#'gender': ['F', 'M', 'M', 'M'],
#'salary': [50000, 70000, 60000, 80000]}
#df = pd.DataFrame(data)
#a. Select rows where age is greater than 30
import pandas as pd
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],'age': [25, 35, 40, 28],'gender': ['F', 'M', 'M', 'M'], 'salary': [50000, 70000, 60000, 80000]}
df = pd.DataFrame(data)
df[df.age>30]
```

```python
#b. Select rows where name contains 'e':

import pandas as pd
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],'age': [25, 35, 40, 28],'gender': ['F', 'M', 'M', 'M'], 'salary': [50000, 70000, 60000, 80000]}
df = pd.DataFrame(data)
df = df[df['name'].str.contains('e')]
print(df)
```

```python
#c. Select rows where gender is 'M' and salary is greater than 65000:

import pandas as pd
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],'age': [25, 35, 40, 28],'gender': ['F', 'M', 'M', 'M'], 'salary': [50000, 70000, 60000, 80000]}
df = pd.DataFrame(data)
df_filtered=df[(df['gender']=='M') & (df['salary']>65000)]
print(df_filtered)
```

```python
#d. Select columns 'name' and 'age':

import pandas as pd
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],'age': [25, 35, 40, 28],'gender': ['F', 'M', 'M', 'M'], 'salary': [50000, 70000, 60000, 80000]}
df = pd.DataFrame(data,columns=['name','age'])
print(df)
```

```python
#e. Select columns 'name' and 'salary' where age is less than or equal to 30:

import pandas as pd
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],'age': [25, 35, 40, 28],'gender': ['F', 'M', 'M', 'M'], 'salary': [50000, 70000, 60000, 80000]}
df = pd.DataFrame(data,columns=['name','age','salary'])
print(df[df['age']<=30])

```
